```python
rohit={
    "Player":[],
    "Runs":[],
    "Opponent":[]
}
for i in df.index:
    if ((p:=df.at[i,"player"]) == "Rohit Sharma"):
        print(p,(r:=df.at[i,"runs"]),(ot:=df.at[i,"opponent"]))
        rohit["Player"].append(p)
        rohit["Runs"].append(r)
        rohit["Opponent"].append(ot)
```

```python
rohit
```

```python
rohitdf=pd.DataFrame(rohit)
```

```python
rohitdf.to_csv("Rohitruns.csv",index=False)
```
